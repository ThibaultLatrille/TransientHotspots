```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(s, b, h):
    xeq = 2*b/(b+1) if h==0.5 else (b-h*s)/(s-2*h*s)
    return np.maximum(0.0, np.minimum(xeq, 1.0))


s_array = np.linspace(6e-6, 0.02, 50)
b_array = np.linspace(6e-6, 0.02, 50)

S, B = np.meshgrid(s_array, b_array)
x = f(S, B, 0.3)

plt.xlabel("s")
plt.ylabel("b")
plt.contourf(S, B, x, 20, cmap='RdGy')
cbar= plt.colorbar();
cbar.set_label("$x_{eq}$")
```
