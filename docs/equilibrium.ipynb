```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(s, b, h):
    xeq = 2*b/(b+1) if h==0.5 else (b-h*s)/(s-2*h*s)
    return np.maximum(0.0, np.minimum(xeq, 1.0))

s_array = np.linspace(6e-6, 0.02, 100)
b_array = np.linspace(6e-6, 0.02, 100)

S, B = np.meshgrid(s_array, b_array)
for h in [0.1, 0.3, 0.4999]:
    x = f(S, B, h)

    plt.contourf(S, B, x, 20, cmap='RdGy')
    plt.gca().yaxis.set_ticks(np.linspace(np.min(B), np.max(B), 10))
    plt.gca().xaxis.set_ticks(np.linspace(np.min(S), np.max(S), 6))
    plt.xlabel("s", fontsize=18)
    plt.ylabel("b", fontsize=18)
    cbar = plt.colorbar()
    if abs(h-0.5) < 0.1:
        plt.title(f"$h \\simeq {h:.1f}$", fontsize=18)
    else:
        plt.title(f"h={h:.1f}", fontsize=18)
    cbar.set_label("$x_{eq}$", fontsize=18)
    plt.tight_layout()
    plt.savefig(f"xeq_{h:.1f}.pdf")
    plt.show()
    plt.close()
```
